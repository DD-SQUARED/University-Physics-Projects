```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define parameters for the binomial distributions
n_values = [5, 20, 20, 20]  # Number of trials
p_values = [3/4, 1/2, 3/10, 1/20]  # Probability of success
labels = [r"$f(5, 3/4)$", r"$f(20, 1/2)$", r"$f(20, 3/10)$", r"$f(20, 1/20)$"]

# Generate x-values (possible number of successes)
x_max = max(n_values)  # Maximum value for x-axis
x = np.arange(0, x_max + 1)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each binomial distribution
for n, p, label in zip(n_values, p_values, labels):
    y = binom.pmf(x, n, p)  # Compute PMF for binomial distribution
    plt.plot(x, y, marker='o', linestyle='-', label=label)

# Labels and title
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial Distributions")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
