```python
# Step 1: Import Required Packages
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Define Constants & Initial Conditions
T0 = 50              # Initial temperature (°C)
T_env = 0           # Ambient temperature (°C)
k = 0.1              # Cooling constant
dt = 0.01             # Time step (seconds)
temp_tolerance = 0.1 # Stop when temperature within this range of T_env
max_time = 1000      # Maximum allowed simulation time (seconds)

# Step 3: Initialize Arrays
T = [T0]             # Temperature history
t = [0]              # Time history

# Step 4: Choose Loop Type (combined stopping condition)
while abs(T[-1] - T_env) > temp_tolerance and t[-1] < max_time:
    # Step 5: Euler Update Rule
    dTdt = -k * (T[-1] - T_env)
    T_next = T[-1] + dTdt * dt

    # Step 6: Append to Arrays
    T.append(T_next)
    t.append(t[-1] + dt)

# Step 7: Plotting Results
plt.plot(t, T, label='Temperature (Euler)')
plt.axhline(T_env, color='r', linestyle='--', label='Ambient Temperature')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title("Newton's Law of Cooling - Euler Method")
plt.legend()
plt.grid(True)
plt.show()

# Optional: print final time and temperature
print(f"Simulation ended at t = {t[-1]:.2f} s with T = {T[-1]:.2f} °C")
```
