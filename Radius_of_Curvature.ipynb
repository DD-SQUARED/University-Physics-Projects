```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
angles = np.array([90, 80, 70, 60, 50])  # Angles in degrees
sin_theta = np.sin(np.radians(angles))  # Sine of the angle
rho_0 = np.array([2.55, 2.25, 2.00, 1.70, 1.20])  # Radius of Curvature in meters

voltages = np.array([150, 200, 250])  # Anode voltages in volts
currents = np.array([1.22, 1.41, 1.59])  # Coil currents in amperes
B = 7.8e-4 * currents  # Magnetic field in teslas, B = (7.8 × 10^-4) × current

# Correct example radius of curvature for different V/B
# Assuming rho_0 for each voltage and B
rho_0_VB = {
    150: np.array([2.55, 2.50, 2.45]),  # Example data for 150 V
    200: np.array([2.25, 2.20, 2.15]),  # Example data for 200 V
    250: np.array([2.00, 1.95, 1.90])   # Example data for 250 V
}

# Graph 1: ρ₀ vs. sin(θ)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(sin_theta, rho_0, 'o-', color='b', label='$\\rho_0$ vs. $\\sin \\theta$')
plt.xlabel('$\\sin \\theta$')
plt.ylabel('$\\rho_0$ (m)')
plt.title('Radius of Curvature vs. Sine of Angle')
plt.grid(True)
plt.legend()

# Graph 2: ρ₀ vs. V/B
plt.subplot(1, 2, 2)
for voltage in voltages:
    VB = voltage / B
    plt.plot(VB, rho_0_VB[voltage], 'o-', label=f'V = {voltage} V')

plt.xlabel('$V / B$')
plt.ylabel('$\\rho_0$ (m)')
plt.title('Radius of Curvature vs. Voltage-to-Magnetic Field Ratio')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
