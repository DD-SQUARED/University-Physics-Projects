```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define Constants & Initial Conditions
m = 1.0             # Mass (kg)
F = 1.0            # Constant force (N)
v0 = 0.0            # Initial velocity (m/s)
x0 = 0.0            # Initial position (m)
dt = 0.01            # Time step (seconds)
t_max = 10          # Maximum simulation time (seconds)

# Step 2: Initialize Arrays
t = [0]
v = [v0]
x = [x0]
a = [F / m]  # Initial acceleration (constant here)

# Step 3: Euler Method Loop
while t[-1] < t_max:
    acceleration = F / m  # Calculate acceleration
    v_next = v[-1] + acceleration * dt  # Update velocity
    x_next = x[-1] + v[-1] * dt          # Update position

    # Append new values to arrays
    a.append(acceleration)
    v.append(v_next)
    x.append(x_next)
    t.append(t[-1] + dt)

# Step 4: Plotting Results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 3)
plt.plot(t, x, label='Position (m)', color='orange')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time')
plt.grid(True)

plt.tight_layout()
plt.show()
```
