```python
#31506534_Andrew Nare

import numpy as np

def statistics(numbers):
    mean_x = np.mean(numbers)
    std_x = np.std(numbers, ddof=1)  # Standard deviation
    sm_x = std_x / np.sqrt(len(numbers))  # Mean standard deviation
    return mean_x, std_x, sm_x

def force_per_unit_length(I1, dI1, I2, dI2, r, dr):
    mu0 = 4 * np.pi * 1e-7  # Vacuum permeability in N/A^2
    r_m = r / 100  # Converting cm to meters

    f = (mu0 * I1 * I2) / (2 * np.pi * r_m**2)

    df_dI1 = (mu0 * I2) / (2 * np.pi * r_m**2)
    df_dI2 = (mu0 * I1) / (2 * np.pi * r_m**2)
    df_dr = -2 * (mu0 * I1 * I2) / (2 * np.pi * r_m**3)

    df = np.sqrt((df_dI1 * dI1) ** 2 + (df_dI2 * dI2) ** 2 + (df_dr * dr / 100) ** 2)  # Convert dr to meters

    return f, df

numbers = [5, 8, 3, 9, 12]
mean_x, std_x, sm_x = statistics(numbers)

I1, dI1 = 30, 4  # Current in Amps
I2, dI2 = 50, 4  # Current in Amps
r, dr = 3.0, 0.5  # Distance in cm

f, df = force_per_unit_length(I1, dI1, I2, dI2, r, dr)

# Round values to 2 significant figures
def round_sig(x, sig=2):
    return float(f"{x:.{sig}g}")

print(f"Mean: {round_sig(mean_x)}, Standard Deviation: {round_sig(std_x)}, Mean Standard Deviation: {round_sig(sm_x)}")
print(f"Force per unit length: {round_sig(f)} N/m ± {round_sig(df)} N/m")
```
